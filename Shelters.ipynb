```python
# Dependencies
import requests
import json
import  gmaps
import pandas as pd
# Google developer API key
from config import (api_key, gkey)
gmaps.configure(api_key=gkey)
# Target city
target_city = "San Francisco, California"

# Build the endpoint URL
target_url = f"https://maps.googleapis.com/maps/api/geocode/json?address={target_city}&key={gkey}"
```

```python
# Run a request to endpoint and convert result to json
geo_data = requests.get(target_url).json()

# Print the json
print(geo_data)
```

```python
# Extract latitude and longitude
lat = geo_data["results"][0]["geometry"]["location"]["lat"]
lng = geo_data["results"][0]["geometry"]["location"]["lng"]

# Print the latitude and longitude
print('''
    City: {0}
    Latitude: {1}
    Longitude: {2}
    '''.format(target_city, lat, lng))
```

```python
target_coordinates = "37.7749295, -122.4194155"
target_search = "Shelter"
target_radius = 1000

# set up a parameters dictionary
params = {
    "location": target_coordinates,
    "keyword": target_search,
    "radius": target_radius,
    "key": gkey
}

# base url
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

# run a request using our params dictionary
response = requests.get(base_url, params=params)
```

```python
# convert response to json
places_data = response.json()

# Print the json (pretty printed)
print(json.dumps(places_data, indent=4, sort_keys=True))
```

```python
name = []
status = []
address = []
lat = []
lng = []
```

```python
for i in range(0, len(places_data['results'])):
    print(places_data["results"][i]["name"])
    print(places_data['results'][i]["business_status"])
    print(places_data["results"][i]["vicinity"])
    print(places_data["results"][i]["geometry"]["location"])
    print("----------------")

    name.append(places_data["results"][i]["name"])
    status.append(places_data['results'][i]["business_status"])
    address.append(places_data["results"][i]["vicinity"])
    lat.append(places_data["results"][i]["geometry"]["location"]["lat"])
    lng.append(places_data["results"][i]["geometry"]["location"]["lng"])
```

```python
shelter_info = pd.DataFrame({
    'Shelter Name' : name,
    'Status': status,
    'Address': address,
    'Lat': lat,
    'Lng': lng
})
shelter_info
```

```python
marker_locations = shelter_info[['Lat', 'Lng']]
fig = gmaps.figure()
markers = gmaps.marker_layer(marker_locations)
fig.add_layer(markers)
fig
```
