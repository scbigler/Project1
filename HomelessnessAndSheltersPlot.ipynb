```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
  

correlation_df = pd.DataFrame([[5669,11,2011], [5895,11,2012], [7008,11,2013], [6408,11,2014], [6775,12,2015],[6996,12,2016],[6858,12,2017],[6857,14,2018],[8035,14,2019],[8124,14,2020],[4000,14,2021]],
                    columns=['homeless', 'shelters', 'year'])

#Scatterplot for the number of homeless shelters each year

 
plt.scatter(correlation_df['year'], correlation_df['shelters'], color='green')
plt.title('Homeless Shelters each Year', fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Number of Homeless Shelters', fontsize=14)
plt.grid(True)
plt.show()
```

```python
#Scatterplot for the number of homeless people each year

plt.scatter(correlation_df['year'], correlation_df['homeless'], color='green')
plt.title('Number of Homeless Each Year', fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Total Number Homeless', fontsize=14)
plt.grid(True)
plt.show()
```
